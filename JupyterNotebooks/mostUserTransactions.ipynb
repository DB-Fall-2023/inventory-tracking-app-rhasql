```python
%matplotlib widget
from flask import Flask, jsonify, request
from flask_cors import CORS, cross_origin
from matplotlib import pyplot as plt
from handler.UserHandler import UserHandler
from DAO.userDAO import UserDAO
from main import app
#import sys
#sys.path.append('http://127.0.0.1:5000/')
```

```python
with app.app_context():
        dao = UserDAO()
        ppJson = dao.getMostUserTransactions()
        mID = []
        mTransactions = []
        result = []
        #print(ppJson)
        for x in ppJson:
            result.append(UserHandler().buildMostUserTransactions(x))
        for y in result:
            mID.append(y['u_id'])
            mTransactions.append(y['Amount of Transactions'])
        plt.figure().set_figwidth(15)
        plt.bar(range(0, len(mTransactions)), mTransactions)
        plt.xticks(range(0, len(mID)), mID)
        # fig, ax = plt.subplots()
        # ax.pie(mTransactions, labels=mID, autopct='%1.1f%%')
        plt.title('Top 3 users with the most transactions')
        plt.show()
        #plt.hist
```
