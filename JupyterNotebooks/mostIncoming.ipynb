Top 5 warehouses with the Most Incoming Transactions

```python
%matplotlib widget
from flask import Flask, jsonify, request
from flask_cors import CORS, cross_origin
from matplotlib import pyplot as plt
from handler.WarehouseHandler import WarehouseHandler
from DAO.warehouseDAO import warehouseDAO
from main import app
#import sys
#sys.path.append('http://127.0.0.1:5000/')
```

```python
with app.app_context():
        dao = warehouseDAO()
        ppJson = dao.getMostIncoming()
        mID = []
        mIncoming = []
        result = []
        #print(ppJson)
        for x in ppJson:
            result.append(WarehouseHandler().buildMostIncoming(x))
        for y in result:
            mID.append(y['warehouse id'])
            mIncoming.append(y['amount of Incoming Transactions'])
        plt.bar(range(0, len(mIncoming)), mIncoming)
        plt.xticks(range(0, len(mID)), mID)
        # fig, ax = plt.subplots()
        # ax.pie(mTransactions, labels=mID, autopct='%1.1f%%')
        plt.show()
        #plt.hist
```
