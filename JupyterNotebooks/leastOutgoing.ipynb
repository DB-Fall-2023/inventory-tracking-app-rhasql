```python
with app.app_context():
        dao = warehouseDAO()
        ppJson = dao.getLeastOutgoing()
        wID = []
        aooTransactions = []
        result = []
        #print(ppJson)
        for x in ppJson:
            result.append(WarehouseHandler().buildLeastOutgoing(x))
        for y in result:
            wID.append(y['warehouse id'])
            aooTransactions.append(y['amount of Outgoing Transactions'])
        plt.figure().set_figwidth(15)
        plt.bar(range(0, len(aooTransactions)), aooTransactions)
        plt.xticks(range(0, len(wID)), wID)
        # fig, ax = plt.subplots()
        # ax.pie(mTransactions, labels=mID, autopct='%1.1f%%')
        plt.show()
        #plt.hist
```
