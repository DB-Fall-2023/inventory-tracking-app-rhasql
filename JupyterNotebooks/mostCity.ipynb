Top 3 warehouses' cities with the most Transactions

```python
%matplotlib widget
from flask import Flask, jsonify, request
from flask_cors import CORS, cross_origin
from matplotlib import pyplot as plt
from handler.WarehouseHandler import WarehouseHandler
from DAO.warehouseDAO import warehouseDAO
from main import app
#import sys
#sys.path.append('http://127.0.0.1:5000/')
```

```python
with app.app_context():
        dao = warehouseDAO()
        ppJson = dao.getMostCities()
        mCities = []
        mTransactions = []
        result = []
        #print(ppJson)
        for x in ppJson:
            result.append(WarehouseHandler().buildMostCities(x))
        for y in result:
            mCities.append(y['warehouse city'])
            mTransactions.append(y['amount of Transactions'])
        plt.figure().set_figwidth(15)
        plt.bar(range(0, len(mTransactions)), mTransactions)
        plt.xticks(range(0, len(mCities)), mCities)
        # fig, ax = plt.subplots()
        # ax.pie(mTransactions, labels=mID, autopct='%1.1f%%')
        plt.show()
        #plt.hist
```
