```python

def getPrice(self):
        dao = PartsDAO()
        ppJson = dao.getPrice()
        pName = []
        pPrice = []
        result = []
        for x in ppJson:
            result.append(PartHandler().buildPprice(x))
        for y in result:
            pName.append(y['p_name'])
            pPrice.append(y['p_price'])
        plt.bar(range(0, len(pPrice)), pPrice)
        plt.xticks(range(0, len(pName)), pName)
        plt.show()
        return jsonify(result)
##PartHandler().getPrice()
```
